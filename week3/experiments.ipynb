```python
import pprint
pp = pprint.PrettyPrinter(indent=4)

while True:
    # Unroll
    query_counts = queries.groupby("category").agg(
                num_queries=('query', 'count')
            ).reset_index().sort_values("num_queries", ascending=False) \
            .assign(
                is_below_threshold_flag=lambda dframe: dframe['num_queries'] < threshold
            )

    # Start from the bottom
    cats_to_unroll = set(query_counts[query_counts["is_below_threshold_flag"] == True]["category"].tolist())

    if not cats_to_unroll:
        break

    # If a category has no parent but needs to be rolledup, just filter out all occurences
    cats_to_filter_out = cats_to_unroll.difference(set(cat_parent_map.keys()))
    
    if cats_to_filter_out:
        queries = queries[~queries["category"].isin(cats_to_filter_out)]
    
    queries["category"] = queries["category"].apply(lambda cat: cat if cat not in cats_to_unroll else cat_parent_map[cat])
```
